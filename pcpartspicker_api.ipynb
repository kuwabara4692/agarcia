```python
# We want to plot a graph of price vs size on monitors from pcparspicker.
# We will connect to pcpartspicker api, select monitors, then create a dataframe of monitors data.
# We will prepare our data(columns price and size) to be plotted, by removing symbols and converting to numeric data type.
from PCPartPicker_API import pcpartpicker
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Show how many pages of monitors are available on pcpartspicker
print("Total monitor pages:", pcpartpicker.lists.total_pages("monitor"))
```

```python
# Assign list of monitors on pcpartspicker to a variable: cpu_info, to easily call that data.
cpu_info = pcpartpicker.lists.get_list("monitor", 2)
```

```python
# Create a dataframe called df1, made of the data from pcpartspicker
df1 = pd.DataFrame(cpu_info)
df1
```

```python
#Look at only price and size of monitors.
for monitor in cpu_info:
    print(monitor["price"], ":", monitor["size"])
```

```python
#Remove $ from column: price.
df1['price']=df1['price'].str.replace('$', '')
print(df1)
```

```python
#Remove double quotes from column: size.
df1['size']=df1['size'].str.replace('"', '')
print(df1)
```

```python
# Check what data types we have for each column in our dataframe. 
# To plot on a graph, we want numeric columns, "object" data type won't work for plotting on graphs.
df1.dtypes
```

```python
# Alternative method, df1.info(), to show data types for our columns in our dataframe.
df1.info()
```

```python
# We will create a second dataframe using columns price and size from our first dataframe.
df2=df1.filter(['price','size'], axis=1)
df2
```

```python
# We will create a 3rd dataframe which will contain the converted columns from df2. 
# Check data types of df3 to for quality assurance.
df3=df2.apply(pd.to_numeric, args=('coerce',))
df3.dtypes
```

```python
# PLotting our df3(price and size now as numeric)
ax=df3.plot.scatter(x='size', y='price')
```

```python
# Using tools from seaborn library(tool: regplot), 
# we are able to use one function to plot our graph and create a best fit line, simultaneously.
sns.regplot(df3['size'],df3['price'])
```

```python
#Save the graph onto your PC. NOTE: the default pathway where the 
#image will be saved is the same folder as this jupyter notebook.
graph1=sns.regplot(df3['size'],df3['price'])
figure=graph1.get_figure()
figure.savefig('pppapi.png', dpi=400)
```
